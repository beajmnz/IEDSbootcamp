```python
# Base ------------------------------------------------------------------
import numpy as np
import pandas as pd

# Association Rules -----------------------------------------------------
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
# Dataset ----------------------------------------------------------------
transactions = [['Bread','Milk','Butter','Jam','Biscuits'],
                ['Bread', 'Milk', 'Beer', 'Chips'],
                ['Jam', 'Soda', 'Milk'],
                ['Soda', 'Chips', 'Beer'],
                ['Chips', 'Bread', 'Beer', 'Milk'],
                ['Jam', 'Beer', 'Soda', 'Milk']]
```

```python
te = TransactionEncoder()
```

```python
# Fitting the transactions
te_fit = te.fit(transactions).transform(transactions)
```

```python
mydata = pd.DataFrame(te_fit, columns = te.columns_)
mydata
```

```python
#Apriori algorithm
apriori_probs = apriori(mydata,min_support=.01, use_colnames=True)

#Association Rules
association_rules(apriori_probs, metric='confidence', min_threshold=.8)
```

```python
association_rules(apriori_probs, metric='lift', min_threshold=2)
```

```python
association_rules(apriori_probs, metric='leverage', min_threshold=.2)
```

```python
association_rules(apriori_probs, metric='conviction', max_threshold=5)
```
