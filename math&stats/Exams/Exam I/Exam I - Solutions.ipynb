## 3.- <font color = "Red"> Probability </font>



### 3.1.- <font color = "Blue"> Depedency (2 Points) </font> 

**For this question you need to split the `RDExpen` and `STJpapers` in three levels denoted by `low`, `mid` and `high` (use the own range of the variables). Then, are the events of `high` level in both variables dependent or independent? Why?**



We can find the probability of intersection $P(\text{high}\cap\text{high})$ from the table of the relative frequencies as

```
pHH = relative_tab.iloc[2,2]
pHH
```

While the probability of each of the *High* can be found as

```
# Number of cases
x_H_expen = science[science['CExpen'] == 'high'].shape[0]
x_H_papers = science[science['CPapers'] == 'high'].shape[0]

# Sample size of each
n_expen = science['CExpen'].dropna().shape[0]
n_paper = science['CPapers'].dropna().shape[0]

# Proportions
p_H_expen = x_H_expen/n_expen
p_H_paper = x_H_papers/n_paper

p_H_expen, p_H_paper
```

Now we can check if

\begin{equation}
P(\text{Exp}_h \cap \text{Paper}_h) = P(\text{Exp}_h)\cdot P(\text{Paper}_h)
\end{equation}

```
pHH == p_H_expen*p_H_paper
```

Since they are not equal we can conclude that they are dependent (to make this more formally we may need an independency non-parametric test)

### 3.2.- <font color = "Blue"> Conditional Probability (2 Points) </font>

**Find the probability that for any randomly chosen country, if the `RDExpen` is not high, the `STJpapers` is high**

We have to look for the following conditional probability

\begin{equation}
P(\text{papers}_h | \overline{\text{exp}_h})
\end{equation}

we may compute it finding each piece from the dataset, but in this case we are going to use Bayes' rule as

\begin{equation}
P(\text{papers}_h | \overline{\text{exp}_h}) = P(\text{papers}_h)\frac{P(\overline{\text{exp}_h} | P(\text{papers}_h)} {P(\overline{\text{exp}_h})} = \frac{P(\text{papers}_h)}{1-P(\text{exp}_h)}\left( 1-\frac{P(\text{papers}_h \cap\text{exp}_h)}{P(\text{papers}_h)}\right)
\end{equation}

just because we have everything from our previous computation

```
prob = p_H_paper/(1-p_H_expen)*(1-pHH/p_H_paper)
print('The probability is: {:4.2%}'.format(prob))
```

which makes sense because there is only one country in the intersection of both variables and exactly the same for the `STJpapers`.

### 3.3.- <font color = "Blue"> Expected Value (2 Points) </font> 

**Assuming that in the world, the research and development expenditure follows a normal distribution with mean 1.5 and standard deviation of 1.1, then find the expected number of countries in the sample that should have an expenditure above 3. Is that number in agreement with what you find in the data? Can you explain?**

```
# Expenditure ------------------------------------------
mean = 1.5
stdev = 1.1

# distribution
distr = ss.norm(mean, stdev)

# probability of >3
x = 3
prob = distr.sf(3)
print('The probability of a country with expenditure greater than 3 is {:4.2%}'.format(prob))

# Number -------------------------------------------------
E = ss.binom(n, prob).mean()
print('The expected number of countries in our sample with an expenditure larger than 3 is {:4.2f}'.format(E))

# Dataset ------------------------------------------------
x_sample = science[science['RDExpen'] > 3].shape[0]
print('The actual number of countries in our sample with an expenditure larger than 3 is {:4.2f}'.format(x_sample))
```
