## <font color = "red"> Cell Phones </font>

**A 2012 survey of  2,254 American adults indicates that 17% of cell phone owners do their browsing on their phone rather than a computer or other device.**

  * **According to an online article, a report from a mobile research company indicates that 38 percent of Chinese mobile web users only access the internet through their cell phones. Conduct a hypothesis test to determine if these data provide strong evidence that the proportion of Americans who only use their cell phones to access the internet is different than the Chinese proportion of 38\%.**
  * **Interpret the p-value in this context**
  * **Calculate a 95% confidence interval for the proportion of Americans who access the internet on their cell phones, and interpret the interval in this context.**

Data given:

 * sample size of 2254
 * sample proportion of 17%
 * proportion of Chinese people 38%

We have to perform a test to see if there is evidence from the sample to determine if American people behaves differently to Chinese people


```
n = 2254
phat = 0.17
P0 = 0.38
```

the decision scheme is

\begin{equation}
H_0:\{P = 0.38\},\quad H_1:\{ P \neq 0.38\}
\end{equation}

We always have to check normality

```
NP = n * phat
NQ = n * (1 - phat)

if NP >= 10:
  if NQ >= 10:
    print("Normality CAN be assumed", NP, NQ)
else:
  print("Normality CANNOT be assumed", NP, NQ)
```

#### <font color = "Blue"> Test Statistic </font>

In this case the condition to reject $H_0$ is

\begin{equation}
\left| \frac{\hat p - P_0}{\sqrt{\frac{\hat p (1 - \hat p)}{n}}} \right| > z_{\alpha/2}
\end{equation}

```
CL = 0.95
SL = 1 - CL

# test statistic
zstat = (phat - P0)/np.sqrt(phat * (1-phat)/n)

# critical value
zcrit = ss.norm.isf(SL)

print(zstat, zcrit)

# test
if abs(zstat) > zcrit:
  print('Reject the NULL hypothesis')
else:
  print('Fail to Reject the NULL hypothesis')
```

Since we Reject the NULL hypothesis, this means that there is enough evidence from our our sample against it (in favor of the alternative). Then our conclusion is that for a 5% of significance level, the proportion of American people who uses only thie cell phones to navigate the internet is not 38%

#### <font color = "Blue"> P-value </font>

```
# p-value
pval = ss.norm.cdf(zstat)

# test
if pval < SL:
  print('Reject the NULL hypothesis', pval, SL)
else:
  print('Fail to Reject the NULL hypothesis', pval, SL)
```

**Test for a 99% of confidence level if the percentage of survivors in each passengers class is greater than 33%** 

To find the number of survivors in each class (total frequency) we do:

 * group_by class
 * select the variable survived
 * sum/count

from there we can find the proportions (relative frequencies)

 * select each of the total frequencies
 * divide by the total nubmer of survivors

```
table_class = titanic.groupby('Pclass')['Survived'].sum() # since it's a variable with 0 and 1, when we sum we are counting how many 1's we have
table_class
```

```
table_class = titanic.groupby('Pclass')['Survived'].count() # with count() we find how many passengers we have in each class, not how many survived
table_class
```

```
table_class = titanic.groupby('Pclass')['Survived'].sum()
table_class
```

We may focus the question from two different perspectives:

 * We may want to test the percentage among the survivors
 * We may want to test the percentage among the number of passengers in each class

Let's go with the first option: In this case the sample size is that of the sum of all the survivors in our previous table, then

### <font color = "Blue"> First Option </font>

Let's go with the first option: In this case the sample size is that of the sum of all the survivors in our previous table, then

```
pclass1 = table_class[1] 
n = table_class.sum() # this sample size is the same for all the classes
phat = pclass1 / n # this is our phat for survivors in 1st class

pclass1, phat
```

We check normality

```
NP = n * phat
NQ = n * (1 - phat)

print(NP, NQ)

if NP >= 10:
  if NQ >= 10:
    print("We CAN use the Normal Approximation since both, NP and NQ are greater than 10")
else:
  print("We CANNOT use the Normal Approximation since NP or NQ are not greater than 10")
```

#### <font color = "Blue"> Test Statistic </font>

Now the test is a right-tailed test so we need to check if

\begin{equation}
\frac{\hat p - P_0}{\sqrt{\frac{\hat p (1-\hat p)}{n}}} > z_\alpha
\end{equation}


```
SL = 0.01
P0 = 0.33

# critical value
zcrit = ss.norm.isf(SL)

# test statistic
zstat = (phat - P0)/np.sqrt((phat * (1 - phat)/n))

if zstat > zcrit:
  print("We Reject the NULL")
else:
  print("We Fail to Reject the NULL")

print(zcrit, zstat)
```

#### <font color = "Blue"> p-value </font>

We may also use the p-value, then

```
# p-value
pval = ss.norm.sf(zstat)

# test
if pval < SL:
  print('Reject the NULL hypothesis', pval, SL)
else:
  print('Fail to Reject the NULL hypothesis', pval, SL)

print(SL, pval)
```

The conclusion is the same: we reject the NULL hypothesis, which implies that the there is evidence from our sample against it and we can conclude, for a 1% of significance level, that the proportion of survivors belonging to the 1st class is greater than the 33% of the survivors.

### <font color = "Blue"> Second Option </font>

Let's now go for the second option, i.e. that the nubmer of survivors in each class was greater than the 33% of the passengers in that class. For this we need to find the sample size as the total number of passengers in each class, then

```
table_n = titanic.groupby('Pclass')['Pclass'].count()
table_n
```

#### <font color = "Blue"> p-value </font>

Let's now do the exercise with the passengers in a different class, let's take the 3rd class

```
pclass3 = table_class[3] 
n = table_n[3] # this sample size is the same for all the classes
phat = pclass3 / n # this is our phat for survivors in 1st class

pclass3, phat
```

Let's assume normality (you must check it) and let's perform the test using the p-value, then

```
# test statistic
zstat = (phat - P0)/np.sqrt((phat * (1 - phat)/n))

# p-value
pval = ss.norm.sf(zstat)

# test
if pval < SL:
  print('Reject the NULL hypothesis', pval, SL)
else:
  print('Fail to Reject the NULL hypothesis', pval, SL)
```

Since now we fail to reject, the conclusion is that there is no evidence against he NULL hypothesis and we conclude that, for a 1% of significance level, the proportion of 3rd class travelers who survived is less than the 33% of all the 3rd class travelers.
