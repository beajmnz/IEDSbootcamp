def roll_dice(N, p1 = 6*[(1/6)], p2 = 6*[(1/6)]):

**Suppose that you have a set of two die such that one of them does not show any even numbers. Comparing both situations, and assuming that you place a bet in the numeric sum. Would you bet for the 7 or for the 6? Have the probabilities changed significantly for the safe die? Simulate the case with a long enough sequence of rolls using the function. Do the results match your computations?**

from general considerations we know that the probability of a 6  comes fromt the following combinations: (1,5), (2,4), (3,3), (4,2) and (5,1), while those for the 7 are (1,6), (2,5), (3,4), (4,3), (5,2) and (6,1). Therefore the probabilities are

\begin{equation}
P(6)=\frac{5}{36} \simeq 0.14, \quad P(7)=\frac{6}{36}=\frac{1}{6} \simeq 0.17
\end{equation}

If one of the die is weighted, say the second, the possible combinations are, for the 6: (1,5), (3,3) and (5,1) with a total number of possible combinations of 18. On the other hand, for the 7 we have (2,5), (4,3) and (6,1)

\begin{equation}
P(6)=\frac{3}{18}=\frac{1}{6}, \quad P(7)=\frac{3}{18}=\frac{1}{6}
\end{equation}

then the probability of a 6 has increased, while that of the 7 remains the same. Let's see how this works in a simulation of a large number of rolls. Then, the function is

````
def roll_die(N, p1 = 6*[(1/6)], p2 = 6*[(1/6)]):
  d1 = np.random.choice([1, 2, 3, 4, 5, 6], size = N, p = p1)
  d2 = np.random.choice([1, 2, 3, 4, 5, 6], size = N, p = p2)
  
  return np.add(d1,d2)
````

### Problem 4. Drugs Test

**In a sport competition, they are using a drug test that detects right the 90% of the cases. If you know from some other researches, that the 8% of the athletes do actually use drugs. Do you think this is a good detection scheme? Make a bar plot with the prior and posterior probabilities.**

Let's write down the two events associated with the problem

\begin{equation}
A:\{\text{athlete uses drugs, athlete doesn't use drugs}\}
\end{equation}

\begin{equation}
B:\{\text{test returns YES, test returns NO}\}
\end{equation}

then the probabilities are

\begin{equation}
P(A_1) = 0.08,\quad P(B_1|A_1) = 0.90,\quad P(B_2|A_2)=0.90
\end{equation}

probably the best way of going through this problem is using Python as a calculator. However, before that we need to knoe what we have to compute.

The answer about a good detection scheme refers to those cases in which the test does not work accurately. In particular we are looking for the probability of a false positive and of a false negative, i.e.

\begin{equation}
\begin{array}{rl}
P(A_1|B_2), & \text{false negative}\\[2ex]
P(A_2|B_1), & \text{false positive}
\end{array}
\end{equation}

which is done using Bayes' formula after we compute the total probabilities $P(B_1)$ and $P(B_2)$. Let's find these probabilities

````
pA1 = 0.08
pA2 = 1- pA1

pB1A1 = 0.90
pB2A1 = 1- pB1A1

pB2A2 = 0.90
pB1A2 = 1- pB2A2

pB1 = pB1A1 * pA1 + pB1A2 * pA2
pB2 = pB2A1 * pA1 + pB2A2 * pA2

print("The probability P(B1) is", str(np.round(pB1, 2)))
print("The probability P(B2) is", str(np.round(pB2, 2)))
````

now we can use Bayes' formula as

````
pA1B2 = pA1*pB2A1/pB2
pA2B1 = pA2*pB1A2/pB1
pA1B1 = pA1*pB1A1/pB1
pA2B2 = pA2*pB2A2/pB2

print("The probability of a TRUE POSITIVE is", str(np.round(pA1B1, 2)))
print("The probability of a TRUE NEGATIVE is", str(np.round(pA2B2, 2)))
print("The probability of a FALSE POSITIVE is", str(np.round(pA2B1, 2)))
print("The probability of a FALSE NEGATIVE is", str(np.round(pA1B2, 2)))
````

Where we have used Bayes Formula in all the cases just to see that they are complementary events.

Now we have to make the bar plot comparing $P(A_1)$ with $P(A_1|B_2)$ and $P(A_1|B_1)$, and $P(A_2)$ with $P(A_2|B_1)$ and $P(A_2|B_2)$. For this part we need the `matplotlib` package, then

````
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('seaborn')
````

and now we use the `plt.bar` function. However it is much easier if we reorganize the probabilities in a data frame. Let's see

````
probs = pd.DataFrame({'Prior': [pA1, pA2],
                      'False Detection': [pA1B2, pA2B1],
                      'True Detection': [pA1B1, pA2B2]},
                     index = ['Use', 'Do not Use']) 

probs.plot.bar()

plt.title('Athletes Drugs\' Test', fontsize = 20)
plt.xlabel('Athlete')
plt.ylabel('Probability')
plt.xticks(rotation = 0)

plt.show()
````

The events we have are the following

\begin{equation}
A:\{\text{construct bp, do not construct bp}\},\quad B:\{\text{pass, fail}\}

\begin{equation}
P(A_1) = 0.80,\quad P(A_2) = 0.20
\end{equation}

\begin{equation}
P(B_1\vert A_1) = 0.86,\quad P(B_1\vert A_2) = 0.65
\end{equation}

from these values, we can write the following tree diagram

````
from graphviz import Graph

g = Graph(node_attr = {'color': 'lightblue2', 'style': 'filled', 'fontsize': '9'},
          edge_attr = {'fontsize': '9'})

g.edge(' ', 'P(A1)', label = " 0.8")
g.edge(' ', 'P(A2)', label = " 0.2")
g.edge('P(A1)', 'P(B1|A1)', label = " 0.86")
g.edge('P(A1)', 'P(B1|A2)', label = " 0.65")
g.edge('P(A2)', 'P(B2|A1)', label = " 0.14")
g.edge('P(A2)', 'P(B2|A2)', label = " 0.35")

g
````

now we have to compute, using Bayes' formula, the probability

\begin{equation}
P(A_1\vert B_1)
\end{equation}

\begin{equation}
P(B_1) = P(B_1|A_1)\cdot P(A_1)+P(B_1|A_2)\cdot P(A_2)

### Problem 2. Defective Items

**A company receives shipments of 40 items. The quality control policy is to inspect a random sample of 10 of the items and if no more than 1 is defective nothing else will be checked.**

  * **Find the probability that a shipment with 7 defective items will receive further checking**
  * **Find the probability that a shipment with 4 defective items is accepted**
  
In this case we have a hypergeometric distribution where $N=40$ and $n=10$. The extra piece is that the shipment is accepted if $k$ is $1$ at most, while for each of the questions $K=7$ and $K=4$, then let's load the `hypergeom` module

````
from scipy.stats import hypergeom
````

and remember from the notes, the order of the arguments in the function call:

```python
hypergeom.pmf(k, N, K, n)
```

then, for the first case the diea is that we impose a posibility of 7 defective items. If we want to know if there is going to be further cheking, we need a value for $k\geq 1$, since with $k=1$ there is no further action, then we look for

\begin{equation}
P(X>1)=S(1)
\end{equation}

the survival probability at 1

````
p = hypergeom.sf(1,40,7,10)

print("The probability that there will be further cheking with 7 defective units is", str(np.round(p,4)))
````

In the second question the shipment is accepted, which means that $k\leq 1$ which means that we look for

\begin{equation}
P(X\leq 1) = F(1)
\end{equation}

the cumulative probability at 1

````
p = hypergeom.cdf(1,40,4,10)

print("The probability that the shipment is accepted with 4 defective units is", str(np.round(p,4)))
````

````
from scipy.stats import poisson
from scipy.stats import expon
````

\begin{equation}
\lambda = 2.5\cdot\frac{5}{1}= 12.5

````
p = poisson.cdf(49, 75)

print("The probability of less than 50 hits each 30 minutes is", str(np.round(p, 5)))
````

````
from scipy.stats import norm
````

````
dist = norm(77, 5)
````

````
p = dist.sf(83)

print("The probability of observing 83ºF or more is", str(np.round(p,4)))
````

````
v = dist.ppf(0.10)

print("The 10% lower temperatures are equal or smaller to", str(np.round(v,4)) + "ºF")
````

\begin{equation}
E[C]=\frac{5}{9}\left(E[F] - 32\right)=25ºC

````
Q1 = dist_c.ppf(0.25)
Q3 = dist_c.ppf(0.75)

IQR = Q3 -Q1

print("The IQR is", str(np.round(IQR, 4)))
````

````
import numpy as np
import pandas as pd
import scipy.stats as ss
import matplotlib.pyplot as plt

from scipy.stats import norm
````

There are two ways of facing this problem. We can give a range and find the pdf directly with that range as could be

```python
x = np.arange(-4,14,0.01)
dist = norm.pdf(x, 5, 3)
```

and then plot it or we may create the instance of a normal distribution and then use it as the base for the rest of the computations. Let's follow this second strategy

````
dist = norm(5, 3)
````

The previous instruction just defines the normal distribution object. Now we can use the different methods with it. Then, for example, the plot can be done as

````
x = np.arange(-4,14,0.01)
plt.plot(x, dist.pdf(x))
plt.show()
````

To generate a random series from the normal distribution we could either use

```python
np.random.normal(5, 3, 1000)
```

or use the instanciated object and the `rvs` method

which means that we have to find the percentiles such that those cumulative probabilities are satisfied with the `ppf` method, then

````
upper = dist.ppf(0.975)
lower = dist.ppf(0.025)

print("The interval containing the 95% of the values around the mean is [", 
      str(np.round(lower,3)), 
      ",", 
      str(np.round(upper,3)), 
      "]")
````

### Problem 8. Supermarket Checkout

**Mark has found that the time it takes to checkout in his local supermarket ranges equally likely from 3 to 10 minutes. If he has just arrived,**

  * **find the expected time he will spend queuing and its standard deviation.**
  * **What is the probability that he will spend more than 8 minutes?**

**If he has already been there for 5 minutes,**

  * **what is the probability that it will take him a a total time less than 7 minutes?**

This is a uniform distribution case since all the events are equally likely, then let's load the module

````
from scipy.stats import uniform
````

remember that we need the arguments `[loc, loc+scale] = [3, 10]` in order to have the non-standarized dstribution. To find the mean and the standard deviation we use

````
m = uniform(loc = 3, scale = 7).mean()
sd = uniform(loc = 3, scale = 7).std()

print("The average minutes he will spend are ", str(np.round(m ,4)))
print("The standard deviation of that time is ",str(np.round(sd ,4)))
````

Then for the next question we have a survival probability as

````
p_8 = uniform.sf(8, loc = 3, scale = 7)

print("The probablity of staying more than 8 minutes is", '{:=2.4}'.format(p_8))
````

For the last question we have the probability

\begin{equation}
P(x<7\vert x>5) = \frac{P(x<7 \cap x>5)}{P(x>5)}=\frac{P(5<x<7)}{P(x>5)}
\end{equation}

then we can find it as

````
prob = (uniform.cdf(7, loc = 3, scale = 7)- uniform.cdf(5, loc = 3, scale = 7))/uniform.sf(5, loc = 3, scale = 7)

print("The probability of statying a total time less than 7 minutes is", str(np.round(prob, 4)))
````

### Problem 9. Commercial Launch

**A company is planning a new commercial launch. According to previous experiences on related products, they estimate that the probability of selling it among the 60%, the 45%, the 25%, the 15%, the 5% and the 1% of their customers is the one given on the table below**


|   Sale’s Probability   | 1% | 3% | 10% | 16% | 25% | 45% | 
| -----------------------|----|----|-----|-----|-----|-----|
|Percentage of customers |60% |45% | 25% | 15% |  5% |  1% |


**To gain some further information, they choose 30 of their customers and ask them whether or not they would buy it, resulting in 4 positive answers. With this new information, ﬁnd out the probability of selling the product to at least the 25% of their customers.**

Let's write the random variable $X$ as

\begin{equation}
X = \{\text{percent of customers who would buy the product}\}
\end{equation}

Then, without the questionnaire, the probability we look for (given the values in the table) would simply be

\begin{equation}
P(x\geq 25\%) = 0.01 + 0.03 + 0.10 = 0.14
\end{equation}

Now, with the information from the questionnaire, the probability we are looking for is

\begin{equation}
P(x\geq 25\% \vert 4)
\end{equation}

i.e. the probability of selling to more than a 25% if 4 people answered yes in the questionnaire. The conditional probability that 4 people would have said "yes" in the questinnaire can be written as

\begin{equation}
P(4\vert X_i) = \pmatrix{30 \\ 4}p_i^4(1-p_i)^{26}
\end{equation}

where the different $p_i$ are read from the table in the problem. Which can be written as

\begin{equation}
P(x\geq 25\% \vert 4) = P(x=25\% \vert 4)+P(x=45\% \vert 4)+P(x=60\% \vert 4)
\end{equation}

and each of these probabilities is found using the Bayes' formula, for example

\begin{equation}
P(X=25\% \vert 4) = \frac{P(X=25\%)\cdot P(4\vert x=25\%z)}{P(4)}
\end{equation}

while $P(4)$ is found using the total probability theorem. Let's find the different probabilities from the `binom` module

````
from scipy.stats import binom

import pandas as pd
````

This problem may be easly done using `for` loops, but we are going to do it step by step. First let's compute the total probability $P(4)$. For that let's write the data table as a dataframe

````
probs = pd.DataFrame({"prob": [0.01, 0.03, 0.10, 0.16, 0.25, 0.45],
                      "custo": [0.60, 0.45, 0.25, 0.15, 0.05, 0.01]})
````

the total probability reads

\begin{equation}
P(4)=P(4\vert 1\%)\cdot 0.45 +\dots+P(4\vert 60\%)\cdot 0.01
\end{equation}

then

````
p4 = (binom.pmf(4, 30, probs["prob"][0])*probs["prob"][0] + 
      binom.pmf(4, 30, probs["prob"][1])*probs["prob"][1] + 
      binom.pmf(4, 30, probs["prob"][2])*probs["prob"][2] + 
      binom.pmf(4, 30, probs["prob"][3])*probs["prob"][3] + 
      binom.pmf(4, 30, probs["prob"][4])*probs["prob"][4] + 
      binom.pmf(4, 30, probs["prob"][5])*probs["prob"][5])

print("The total probability P(4) is ", str(np.round(p4,5)))
````

And the probability we look for is

````
p_25more = (binom.pmf(4, 30, probs["prob"][0])*probs["prob"][0] + 
            binom.pmf(4, 30, probs["prob"][1])*probs["prob"][1] +
            binom.pmf(4, 30, probs["prob"][2])*probs["prob"][2])/p4

print("The conditional probability is ", str(np.round(p_25more, 5)))
````

Which makes a significant change from the prior probability of 0.14

### Problem 10. Triathlon

**In triathlons, it is common for racers to be placed into age and gender groups. Friends Leo and Mary both completed the Hermosa Beach Triathlon, where Leo competed in the Men, Ages 30 - 34 group while Mary competed in the Women, Ages 25 - 29 group. Leo completed the race in 1:22:28 (4948 seconds), while Mary completed the race in 1:31:53 (5513 seconds). Obviously Leo finished faster, but they are curious about how they did within their respective groups. Can you help them? Here is some information on the performance of their groups:**

  * **The finishing times of the Men, Ages 30 - 34 group has a mean of 4313 seconds with a standard deviation of 583 seconds.**
  * **The finishing times of the Women, Ages 25 - 29 group has a mean of 5261 seconds with a standard deviation of 807 seconds.**
  * **The distributions of finishing times for both groups are approximately Normal.**

**Remember: a better performance corresponds to a faster finish.**

  * **Write down the short-hand for these two normal distributions.**
  * **What are the Z scores for Leo's and Mary's finishing times? What do these Z scores tell you?**
  * **Did Leo or Mary rank better in their respective groups? Explain your reasoning**
  * **What percent of the triathletes did Leo finish faster than in his group?**
  * **What percent of the triathletes did Mary finish faster than in her group?**
  * **If the distributions of finishing times are not nearly normal, would your answers to parts (2) - (5) change? Explain your reasoning.**
  * **Under the assumption of normality, what is the cutoff time for the fastest 5% of athletes in the men's group, i.e. those who took the shortest 5% of time to finish.**
  * **What is the cutoff time for the slowest 10% of athletes in the women's group?**


The distributions we have are $X\sim N(4313, 485)$ for men and $Y\sim N(5261, 807)$ for women. Then if the times of Leo and Mary are 4948 and 5513 respectively, their z-scores are

\begin{equation}
Z_{\text{Leo}}=\frac{4948-4313}{485}
\end{equation}

and

\begin{equation}
Z_{\text{Mary}}=\frac{5513 - 5261}{807}
\end{equation}

````
m_men = 4313
m_women = 5261

sd_men = 485
sd_women = 807

t_leo = 4948
t_mary = 5513

Z_Leo = (t_leo - m_men)/sd_men
Z_Mary = (t_mary - m_women)/sd_women

print('The Z-Score of Leo is {:<6.3f}, while that of Mary is {:<6.3f}'.format(Z_Leo, Z_Mary))
````

Now, with respect to their own groups, Leo performance is worst since it is further from its mean than Mary's time.

To find the percents of triathletes we use the survival probability of the normal distribution in the z-scores found before, since we are asked for the triathletes who were slower than Leo and Mary, then

````
p_leo = norm.sf(Z_Leo)
p_mary = norm.sf(Z_Mary)

print('The percent of triathletes slower than Leo is {:<4.2%}, and that slower than Mary is {:<4.2%}'.format(p_leo, p_mary))
````

The cutoff time for the 5% fastest atheletes in each category is given by the percentile function, same for the 10% slowest (note that we could do it with the survival functions too), then

````
fastest_men = norm.ppf(0.05, m_men, sd_men)
fastest_women = norm.ppf(0.90, m_women, sd_women)

print('The cutoff time for the 5% fastest men is {:<4.2f} seconds, and that for the 10% slowest women is {:<4.2f} seconds'.format(fastest_men, fastest_women))
````
