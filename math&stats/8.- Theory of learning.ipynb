# Theory of Learning

## Introduction

In this chapter we begin now a journey into a new realm which is to some extent more than an application of our previous notions on probability and statistics, but which cannot be understood without them: that of **Statistical Learning**. ote that when we speak about learning we are just dealing with **dependency estimation problems**, i.e. those situations in which we perform an estimation based on some data. 

It is true that if we just define our learning as we have just made it does not look like we are dealing we a new set of ideas. However, as we will see, we are dealing with a new paradigm that was developed from the 1960s thanks to the appearance of computers which allowed the analysis of real-life multidimensional models with large datasets. The main first problem that resulted from this type of analysis is what Bellman (1961) denoted *the curse of dimensionality*, namely, the idea that when the dimensions of the feature space increase, the data may become more sparsed and then we face the need of larger amounts of data (exponentially growing) to make reliable and sound estimations.

The search for a solution to this problem is where we find the seeds of this new paradigm. However, this field has evolved in many directions from the first problems of pattern recognition by extending and developing techniques which now receive different names as *statistical learning*, *machine learning*, *deep learning* and so. Not all of the refer to the same concepts since they differ mostly in the techniques used and in the overall aims and objectives. Then, for example, machine learning makes a special emphasis in predictions while statistical learning puts it on inference, reason why there is a usual claim that goes by saying that machine learning is data plus computable structures while statistical learning is data plus analytical theory.

We will refer to the general process of learnign as a process of **Data Analytics** which means that we will use data to draw conclusions from the information contained within it, a process that can fall in any of the following three categories:

 * **Descriptive**, where through the use of **Data Mining** and **Data Aggregation**, we will use the data to answer the question of "what has happened?"
 * **Predictive**, where through the use of **Statistical Models** and **Forecasting Techniques** we try to understand what may happen;
 * **Prescriptive**, where through the use of **Optimization** and **Simulation** algorithms we try to gain insights to advice on possible outcomes and then decide what to do


## Statistical Models

To put the previous ideas in a more concrete context, we should say that all our learning task deals with the estimation of a *target function* that determines a **Statistical Model**. However, there are different perspectives when we approach this problem. The general scheme can be seen as

$$
\text{Input} \longrightarrow \text{Model}\longrightarrow \text{Output}
$$

then we classify our task with respect to what we try to find in that scheme:

 * **Optimization**, when we know the model and the output, then we look for an Input that used in the model produces the output
 * **Modelization**, when we know the input and output sets, then we look for the model that returns the closest output for every known input
 * **Simulation**, when we know the input set and the model, then we obtain the outputs by supplying the inputs to the model

In this context we are always assuming that the model is directional , however, the tasks above are completely different and with respect to the assumption and interrelated, for example, finding a simulator implies solving a modelling problem; any modelization problem can be turned into an optimization one by requiring that the *Error Rate* is minimum or the *Hits Rate* maximum. This is a key point for all the Statistical Decision Theory. 

We can define the **Search Space** as the collection of objects of interest (including the solution) of the problem and it contains the inputs, the models and the outputs. The dimensionality of this space will give us one hint on the complexity of the model.

In the inputs we find the set of variables typically denoted as $X$ and known as *predictors*, *independent variables*, *features* or *exogenous variables* (to mention some of the names). On the other hand, in the output we find the variables denoted as $Y$ and known as *response*, *dependent* or *endogenous* variable(s). Then the modelling pipeline above can be formally written as

$$
Y = f(X)+u
$$

where $f()$ denotes the model itself, i.e. the systematic part of the response that can be explained using the variables $X$, as a fixed but yet unkown function, and $u$ are the random errors where we must include not only the *confounding variables* but also the own noise of our random variables. 

the process of estimation or modelling is that of finding a best fit functional form for $f(X)$ which will always be dependent of the data set we have. The main reasons behind estimation are **prediction** and **inference**.



### Prediction vs. Estimation

We talk about **prediction** when we want to use the estimated model

$$
\hat Y = \hat f(X)
$$

to obtain particular values of the response variable. A rather common example may be that of predicting the income of a person in terms of the years of education.

In this sense we must understand that, in general, the estimated value will be different from the measured value for the same $X$ and even more, both of them will be different from the actual true value which we will never know. This allow us to compute the MSE of the model as

$$
E[(Y-\hat Y)^2]=[\text{bias}(\hat f(X)]^2+Var(\hat f(X))+Var(u)
$$

which is splitted in an irreducible term defined by the variance of the error term and another reducible written as the MSE of the estimator defined by the model. In the context of Statistical Decision Theory, this expected value is known as the **expected predition error** (EPE) and is the expected value of the **loss function**.

It is the minimization of this EPE function is what leads to the different models we will see in this chapter and in general all the models try to minimize the reducible part of the function.

If we are not interested in predicting values of the response but on seen the way in which $Y$ is affected by the different $X$ we talk about **inference** of the model. It is in this context where all the notions of hypothesis testing come to our rescue: which predictors are associated to the response? Which set of variables can we use to describe the behavior of the response? What is the individual relation between predictors and response?



### Parametric vs. Non-parametric Estimation

The estimation processes can be either **parametric** or **non-parametric**, refer to whether or not we will make an explicit assumption on the model for $f(X)$ and then try to estimate it. 

For example, when we perform a *linear* or a *logistic regression*, we assume that the functional form of the response as a function of $X$ is that of a linear function

\$$
Y=\beta_0+\beta_i\,X_i
$$

Then, the main parametric procedure is that of **least squares** that we will study shortly.

In other cases, as for example using *splines* are non-parametric since they do not assume any specific functional form. 

Each procedure has its own pros and cons that are somewhat far from the scope of the course. There is, however, one potential problem with parametric models and it is that since we do not really know the population model, the estimated model may not follow the actual true values. The idea is that in general they are designed to capture a particular property of the population through a simplication of the true overall behavior.



### Supervised vs. Unsupervised Learning

In our previous discussion we have implicitely assumed that we know the response variables. However, that is not always the case, a for example in a market segmentation study, where we split the population according to some of their properties: age, income, zip code or any other, but we want to understand the spending habits of that population. These last partterns will never be known for us.

If we know the response variable we talk about **supervised learning**, in the sense that the response let us see how close to it we are with our model and then, in some sense it is supervising us. However, if we do not know it, we speak of an **unsupervised learning**. In this context, the regression and classification linear models we will study are always supervised procedures.



### Regression vs. Classification Models

Another important differentiation in the models is that of regression and classification. In this case we are making reference to the nature of the response variable. If the response variable is quantitative, the model will be a {\bf{regression}}, but if the response is qualitative, the model will be a {\bf{classification}}, then if we want to study how gender affects income, that will be a regression problem. However, if we want to see if we may determine the gender from the income, we will be classifying the population.



## Validation of the Model

Up to this point we have an idea of the general procedure to estimate a learning model: choose a model, then its hyperparameters, fit it and make some predictions. 

In this section we deal with the election of the hyperparameters, which belongs to the process generally known as **validation**. The idea of this procedure is that we need to be sure that both, the model we choose and the values of the parameters we use are the right ones to make a good fit of the data. The validation procedure is understood as the **model selection**, i.e. the procedure in which we estimate the performance of different models by including or excluding variables and through the study of all the potential problems that may arise in the training (outliers, heteroskedasticity,.... On the hand, the testing is understood as the **model assessment** where, once found the final model we assess the model by estimating the error on the test data set.

The strategy we follow is that of splitting our sample in three sets

\begin{equation}
\text{Training} \longrightarrow \text{Validation} \longrightarrow \text{Test}
\end{equation}

why? because we want to be sure that the model works well off-sample, i.e. in a dataset that is not the one used to generate it. In order to generate these sets from our sample we will use `scikit-learn` in Python via the function `train_test_split()`. 

There are two main strategies to follow in this context: **hold-out sets** and **Cross-validation**



### Hold-out sets

In this case we make a single split of our data and use each part for what we need. A usual rule is using 60%/20%/20%, for train/validation/test, but these values may change according to the circumstances.

```
# Modules -------------------------------------------
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

Let's load a dataset and generate the holdout sets

```
# Dataset -------------------------------------------
iris = load_iris()
```

If you want to see a dataframe with the dataset you may want to import it from `seaborn`. Now, the target in this dataset is the `species`, then let's generate the features and dependent variables from this dataset

```
X = iris.data
Y = iris.target
```

Now we can find the holdout sets as

```
X1, X2, y1, y2 = train_test_split(X, Y, 
                                  random_state = 1,
                                  train_size=0.5)
```

in this code we have generated two sets from the features and another two from the target, each with a size of 50% of the sample. Let's find the accuracy of a KNN classifier found from the first set, and tested on the second

```
# Model ----------------------------------------------
model = KNeighborsClassifier() # We leave the hyperparameters by default

# Fit ------------------------------------------------
model.fit(X1, y1)

# Prediction and Accuracy ----------------------------
Y2_hat = model.predict(X2)
accuracy_score(y2, Y2_hat)
```

```
model
```

### Cross-validation

The previous procedure is better than finding the accurary with the same set used to fit the model but it still has a major drawback: it is too dependent on the split (try changing the random seed in the code to see the differences in the score).

To solve this problem we use the strategy of cross-validation, where we perform the split the dataset in a sequence of different foldings and use *all* to fit different models. The final accuracy score can be found, say, finding the average of all the models.

For example, in the case before we may have used $(y2, X2)$ for the fit and $(y1, X1)$ for the test, then




```
# Models ----------------------------------------------------
model_1 = model.fit(X1, y1)
model_2 = model.fit(X2, y2)

# Fits ------------------------------------------------------
y2_hat = model_1.predict(X2)
y1_hat = model_2.predict(X1)

# Predictions and Accuracies --------------------------------
score_1 = accuracy_score(y2, y2_hat)
score_2 = accuracy_score(y1, y1_hat)
print(score_1, score_2)
(score_1 + score_2)/2
```

this procedure we have used is known as **two-fold cross-validation**, and can be generalized to any number of foldings. The extreme case is the **Leave out one** folding where the test is made with samples of size 1 and then we have as many foldings as our sample size.

Of course there is no need to take the manual approach of the previous code. We have a function for it

```
from sklearn.model_selection import cross_val_score
```

now we can use it for a **five-fold cross validation** as

```
scores = cross_val_score(model, X, Y, cv = 5)
scores.mean()
```

## Model Selection

Once we make a difference in the training, validation and test sets, we can see that the idea is that the model built from the training data should be accurate enough once we use it in previously unseen data. If we have, for example, built a model from clinical measurements in a study for the prediction of diabetes or to infer their relation with diabetes, we want to be able to use it with future patients. Mathematically we want a model that gives the lowest **MSE** for the test set, not for the training set.

Since the MSE has contributions coming from the bias and the variance of the model, finding the best model is equivalent to finding the *right* spot in the **trade-off between bias and variance** (since we cannot make both at the same time as small as we may want)


### Bias-Variance Trade-off

From the equation for the EPE, we see that the way to minimize it is by choosing a method that produces both, a low bias and a low variance. 

 * The **variance** we are interested on is the amount by which the estimated model would change if we use a different training set, 
 * The **bias** refers to the error introduced by choosing an approximation (function of the model) to the real life problem.

It is relatively easy to deal with methods that give a low value of the bias or of the variance, but not together.  

In flexible models, the variance will increase and the bias will decrease, i.e. we will be closer to the values but in a very unstable way (this is the case of **K-nearest neighbourhood method**). On the other hand with restricted models, we obtain models with low variance but large bias, i.e. maybe not close to the points but very stable (this is the case of the **linear regression**). 

Said differently, a more complicated model will give worse results, since the model will keep on improving the train scores while underperform in the test set (**high-variance**). On the other hand, if we move towards a simpler model we may underperform in both sets (**high-bias**)

Thinking that a flexible model may overperform another simpler and more restricted would be an error. In fact it is not unsual to find the contrary situation, a counterintuitive situation that arises due to the **overfitting** problem. This is the case in which the training MSE decreases significantly but only because our model is finding some patterns due to random effects and not to the true behavior, i.e. it is following the white noise of the model. 

On the other hand, we find **underfitting**, when the model is not complex enough to capture the true dependency of the target based on the features.

In order to analyse these behaviours we have two different graphical approaches: the **Learning** and the **Validation** curves



### Validation Curve

It is a diagnosis curve for the model that measures the changes in the accuracy with respect to any of the hyperparameters of the model. In the representation we always find a training and a cross-validation curve.

We must emphasize that this curve is not used to fit the model since we may fall into a bias towards the training set and then into a problem with overfitting.

With respect to its interpretation, it is important to notice that we will have **overfitting** if the training curve grows too fast and leaves the cross-validation curve behind. On the other hand, **underfitting** occurs when the training curve grows too slowly and goes behind the cross-validation one, roughly when both scores are low.

```
from sklearn.model_selection import StratifiedKFold, GridSearchCV, ShuffleSplit
from sklearn.model_selection import validation_curve
import matplotlib.pyplot as plt
```

In our KNN example, we are going to use the number of neighbors (from 1 to 10), then

```
# Validation curve -------------------------------------------------------------
param_range = np.arange(1, 10, 1)
train_scores, test_scores = validation_curve(model, 
                                             X, 
                                             Y, 
                                             cv=5, 
                                             scoring="accuracy",
                                             param_range = param_range,
                                             param_name = 'n_neighbors')

# Mean Values ------------------------------------------------------------------
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)

# Graph ------------------------------------------------------------------------
plt.figure(figsize = (12,5))
plt.ylim(0.9, 1.0)

plt.plot(param_range, train_scores_mean, label="Training score",
             color="darkorange")
plt.fill_between(param_range, train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std, alpha=0.1,
                 color="darkorange")
plt.plot(param_range, test_scores_mean, label="Cross-validation score",
             color="navy")
plt.fill_between(param_range, test_scores_mean - test_scores_std,
                 test_scores_mean + test_scores_std, alpha=0.1,
                 color="navy")

plt.title("KNN - Validation Curve", fontsize = 18)
plt.xlabel('Number of Neighbors', fontsize = 15)
plt.ylabel("Accuracy Score", fontsize = 15)


plt.legend(loc="best")
plt.show()
```

in this case it looks like for a low number of neighbors, the model may be overfitting (high training score and low cross-validation), otherwise the model performs well. To be fair, the difference at the beginning is not that much as to declare overfitting, but with a larger range, that may be the case.

### Learning Curve

This curve gives the relation between the cost and the output of the model. The cost in our case will be related with the computational time and with the value of the sample size used to train the model. In the graph we see two different lines:

 * The **training learning curve**, which measures how well is the model learning with an increase of the cost.
 * The **cross-validation curve**, which tell us how well the model generalizes with an increase of the cost

In this context we say that there is **underfitting** if the training curve is not able to grow (the model cannot learn) or if the cross-validation vurve is constantly decreasing until the end. On the other hand, we say that there is **overfitting** if the training curve is constantly decreasing or if the validation curve decreases to increase at the end.

Let's find it

```
from sklearn.model_selection import learning_curve
import numpy as np
```

Now we are going to plot two curves, the learning curve itself and the scalability curve, which reflects the computational cost of the model with an increase in the sample size.

```
# Learning Curve ---------------------------------------------------------------
cv = ShuffleSplit(n_splits=100, test_size=0.2, random_state=0)
train_sizes, train_scores, test_scores, fit_times, _ = learning_curve(model, 
                                                                      X, 
                                                                      Y, 
                                                                      cv = cv, 
                                                                      n_jobs = -1,
                                                                      train_sizes = np.linspace(.1, 1.0, 10),
                                                                      return_times = True)

# Mean Values ------------------------------------------------------------------
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)
fit_times_mean = np.mean(fit_times, axis=1)
fit_times_std = np.std(fit_times, axis=1)


plt.figure(figsize = (18,8))
plt.suptitle("KNN - Learning Curve", fontsize = 25)

# Learning Curve ---------------------------------------------------------------
plt.subplot2grid((1,2), (0,0))
plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                         train_scores_mean + train_scores_std, alpha=0.1,
                         color="r")
plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                         test_scores_mean + test_scores_std, alpha=0.1,
                         color="g")
plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
                 label="Training score")
plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
                 label="Cross-validation score")
plt.title("Learning Curve", fontsize = 18)
plt.xlabel("Training examples", fontsize = 15)
plt.ylabel("Accuracy Score", fontsize = 15)
plt.legend(loc="best")

# Scalability Plot -------------------------------------------------------------
plt.subplot2grid((1,2), (0,1))
plt.plot(train_sizes, fit_times_mean, 'o-')
plt.fill_between(train_sizes, fit_times_mean - fit_times_std,
                         fit_times_mean + fit_times_std, alpha=0.1)
plt.title("Scalability of the model", fontsize = 18)
plt.xlabel("Training examples", fontsize = 15)
plt.ylabel("Fitting Time (sec)", fontsize = 15)


plt.tight_layout(rect = (0,0,0.95,0.95))
plt.show()
```
