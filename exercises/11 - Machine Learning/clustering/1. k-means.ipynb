# Clustering

They are unsupervised Machine Learning techniques in which we do not have a target and in which we want to group the data to find patterns. The idea is to automatically find groupings or clusters of elements according to a measure of similarity between them.

The fundamental objective of clustering techniques is to identify groups or clusters of elements such that:

- The average similarity between elements of the same cluster is high. High intra-cluster similarity.

- The average similarity between elements of different clusters is low. Low inter-cluster similarity.




There are several classes of clustering according to the technique used to separate the groups:
    
    - Hierarchical clustering.
    
    - Partition clustering.

We will focus here on **partition clustering**.

## K-means

The partition clustering technique distributes the elements among a predetermined number of clusters or groups. We define the centroid of a cluster as that element that minimizes the sum of the similarities to the rest of the elements of the cluster: This technique receives as input the number of clusters to be formed in addition to the elements to be classified and the matrix of similarities.


The most popular partition clustering technique is **K-means**. In K-means we define the centroid of a cluster as that element that minimizes the sum of the similarities to the rest of the elements of the cluster.

The steps of the algorithm are as follows:

- **Step 1:**: Initialize the centroids. Standard method is to randomly select input data points as the initial centroids.


- **Step 2:** Create k clusters assigning each element to the closest centroid.


- **Step 3:** Calculate new centroids as those elements that minimize the sum of the distances to the rest of the cluster elements.


- **Step 4:** Return to step 2 while there is change in the clusters or a maximum number of iterations is reached.

## Artificial Data Creation

```python
from sklearn.datasets import make_blobs
?make_blobs
```

```python
X, y_true = make_blobs(n_samples=300, centers=4,
                       cluster_std=0.60, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50);
```

## K-means Model

```python
?KMeans
```

```python
# [1] Import model
from sklearn.cluster import KMeans as model_constructor

# [2] Import metric

# [3] Define model
model = model_constructor(n_clusters = 4)

# [4] Train model
model.fit(X)

# [5] Predict
y_kmeans = model.predict(X)
y_kmeans 

# [6] Compute metric
```

```python
dir(model)
```

## Results Visualization

```python
plt.scatter(X[:, 0], X[:, 1], c = y_kmeans, s=50, cmap='viridis')
centers = model.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

## Elbow Method

```python
wcss = []
for i in range(1, 11):
    model = model_constructor(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    model.fit(X)
    wcss.append(model.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

The optimum number seems to be 3-4
